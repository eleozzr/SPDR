# Data of Simulation 1

This simulated data is related to **Figure 2A and 2C** in the article.

Set parameters: 5000 genes, 4 cell types, 3 batches with batch effect.

```python
%load_ext rpy2.ipython
```

```python
%%R -o counts -o meta

suppressMessages(library(splatter))

params <- newSplatParams()
params <- setParam(params, "nGenes", 5000)
params <- setParam(params, "batchCells", c(1000,1000,2000))
params <- setParam(params, "batch.facLoc", 0.5)
params <- setParam(params, "batch.facScale", 0.5)
params <- setParam(params, "group.prob", c(1/4,1/4,1/4,1/4))
params <- setParam(params, "de.facLoc", 0.3)
params <- setParam(params, "dropout.shape",-1)
params <- setParam(params, "dropout.type","experiment") #The
params <- setParam(params, "seed",28) 
sim <- splatSimulate(params, method="groups", verbose=FALSE)

counts = data.frame(counts(sim))
meta = data.frame(colData(sim))
```

# UMAP visualization for raw data

```python
"""import necessary packages"""
import scanpy as sc 
import anndata 
from scipy.sparse import csr_matrix
```

```python
"""save data"""
adata = anndata.AnnData(X=csr_matrix(counts.values.T), obs=meta)
adata.write("./self_sim_complex.h5ad")
"""preprocessing"""
sc.pp.normalize_total(adata,target_sum=10000)
sc.pp.log1p(adata)
sc.pp.highly_variable_genes(adata, n_top_genes=1000, subset = True)
sc.tl.pca(adata,svd_solver='arpack')
"""umap plot"""
sc.pp.neighbors(adata)
sc.tl.umap(adata)
sc.pl.umap(adata,color=['Batch','Group'])
```
